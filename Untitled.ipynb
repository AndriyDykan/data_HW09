```python
import tensorflow as tf
import numpy as np
import pandas as pd 
from scikeras.wrappers import KerasClassifier 
from sklearn.model_selection import GridSearchCV 


image_data = tf.keras.datasets.fashion_mnist
(train_image,train_label),(test_imeges, test_labels) = image_data.load_data()
train_image = train_image /255
train_image = train_image.astype(float)
test_imeges = test_imeges /255
test_imeges = test_imeges.astype(float)


train_image = train_image.reshape(train_image.shape[0],28,28)
test_imeges = test_imeges.reshape(test_imeges.shape[0],28,28)


#train_label =  tf.keras.utils.to_categorical(train_label,10)
#test_labels =  tf.keras.utils.to_categorical(test_labels,10)
test_labels.shape
```

```python
import keras_tuner 
import keras

def build_model(hp):
    model = keras.Sequential()
    model.add(tf.keras.layers.Flatten(input_shape=(28, 28)))
    
    # Додайте кілька Dense шарів з Dropout
    for i in range(hp.Int('num_layers', 2,15)):
        model.add(keras.layers.Dense(
            units=hp.Int(f'filters_{i}', 32, 320, step=32),
            activation='relu'))
        model.add(keras.layers.Dropout(rate=hp.Float(f'dropout_rate_{i}', 0.1, 0.3, step=0.1)))
    
    # Додайте останній Dense шар для класифікації
    model.add(tf.keras.layers.Dense(units=10, activation='softmax'))
    
    # Вибір оптимізатора
    optimizer_choice = hp.Choice('optimizer', ['adam', 'rmsprop', 'sgd'])
 
    if optimizer_choice == 'adam':
        optimizer = keras.optimizers.Adam()
    elif optimizer_choice == 'rmsprop':
        optimizer = keras.optimizers.RMSprop()
    else:
        optimizer = keras.optimizers.SGD()

    # Компіляція моделі з вибраним оптимізатором та метрикою
    model.compile(optimizer=optimizer,
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    return model

# Ініціалізація Tuner
tuner = keras_tuner.RandomSearch(
    build_model,
    objective='val_loss',
    max_trials=5,
    directory='notebook/9/ker1235'
)

# Пошук найкращої моделі
tuner.search(train_image, train_label, epochs=keras_tuner.HyperParameters().Choice('epochs', [20,30]), validation_data=(test_imeges, test_labels), batch_size=keras_tuner.HyperParameters().Choice('batch_size', [150]))

# Отримання найкращої моделі
best_model = tuner.get_best_models()[0]
test_loss, test_accuracy = best_model.evaluate(test_imeges, test_labels)
```
